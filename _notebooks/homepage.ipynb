<h2>// Object-Oriented Programming</h2>
<p>Building reusable, modular game components through classes, inheritance, and method overriding. The foundation of every character, enemy, and platform in the game.</p>
<div class="topic-grid">
<a class="topic-card" href="https://chetan-coder.github.io/portfolioC/oop/writing-classes"><h3>Writing Classes</h3><p>Create minimum 2 custom character classes extending base classes (Character, Enemy, NPC, Platform).</p></a>
<a class="topic-card" href="https://chetan-coder.github.io/portfolioC/oop/methods-parameters"><h3>Methods & Parameters</h3><p>Implement 5+ methods with parameters and return values (e.g. <code>collisionHandler(other, direction)</code>).</p></a>
<a class="topic-card" href="https://chetan-coder.github.io/portfolioC/oop/instantiation-objects"><h3>Instantiation & Objects</h3><p>Instantiate game objects in GameLevel configuration using Object Literals (data-driven design).</p></a>
<a class="topic-card" href="https://chetan-coder.github.io/portfolioC/oop/inheritance"><h3>Inheritance (Basic)</h3><p>Create class hierarchy with 2+ levels &mdash; e.g. <code>GameObject → Character → Player</code>.</p></a>
<a class="topic-card" href="https://chetan-coder.github.io/portfolioC/oop/method-overriding"><h3>Method Overriding</h3><p>Override parent lifecycle methods: <code>update()</code>, <code>draw()</code>, <code>handleCollision()</code>.</p></a>
<a class="topic-card" href="https://chetan-coder.github.io/portfolioC/oop/constructor-chaining"><h3>Constructor Chaining</h3><p>Use <code>super(data, gameEnv)</code> to chain constructors and forward initialization data.</p></a>

<a class="topic-card" href="https://chetan-coder.github.io/portfolioC/control/conditionals"><h3>Conditionals</h3><p>Use <code>if/else</code> branches to implement collision detection and state transitions.</p></a>
<a class="topic-card" href="https://chetan-coder.github.io/portfolioC/control/nested-conditions"><h3>Nested Conditions</h3><p>Multi-level conditionals for complex game logic (e.g. power-up + collision + direction).</p></a>
