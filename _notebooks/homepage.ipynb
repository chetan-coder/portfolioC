<div class="category">
<h2>// Object-Oriented Programming</h2>

<div class="category-intro">
<p>Object-Oriented Programming is the architectural backbone of every non-trivial game. Instead of writing one massive script with thousands of loose variables, OOP lets us bundle related state (position, velocity, health) and behavior (move, draw, collide) into self-contained classes &mdash; and then reuse them by extending those classes for specific entities. The result is code that's readable, maintainable, and easy to grow without rewriting.</p>
<p>In GateGame v2, every visible entity on the canvas &mdash; the player, the patrolling enemies, the dialogue NPCs, the platforms, the background layers &mdash; is an instance of a class. They all trace back through an inheritance chain to a single <code>GameObject</code> base class provided by the OCS GameEngine. Adding a new enemy type means writing a small subclass that overrides a few methods, not duplicating hundreds of lines of boilerplate.</p>
</div>

<a class="evidence" href="https://chetan-coder.github.io/portfolioC/oop/writing-classes">
  <div class="evidence-head">
    <h3><span class="evidence-num">[01]</span>Writing Classes</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Create minimum 2 custom character classes extending base classes (Character, Enemy, or NPC).</div>
  <p class="evidence-impl">We authored three custom classes &mdash; <code>Player.js</code>, <code>NPC.js</code>, and <code>Enemy.js</code> &mdash; each extending a base class from the GameEngine. Each one encapsulates its own sprite paths, animation frames, default hit box, and unique behaviors. Player handles keyboard input and score, NPC handles dialogue triggers and AI-backed responses, and Enemy handles patrol pathing and damage dealing. Three classes comfortably exceeds the 2+ requirement and gives the game enough variety to feel alive.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/oop/methods-parameters">
  <div class="evidence-head">
    <h3><span class="evidence-num">[02]</span>Methods & Parameters</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Implement 5+ methods with parameters and return values (e.g. <code>collisionHandler(other, direction)</code>).</div>
  <p class="evidence-impl">Each class implements multiple parameterized methods. The most-used signature is <code>collisionHandler(other, direction)</code>: it receives the OTHER object involved and a direction string (<code>top</code>, <code>bottom</code>, <code>left</code>, <code>right</code>) so the correct response fires. Other examples include <code>update(deltaTime)</code>, <code>draw(ctx)</code>, <code>applyForce(fx, fy)</code>, and value-returning helpers like <code>getHitBox()</code> and <code>isOnGround()</code> &mdash; well beyond the 5-method bar.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/oop/instantiation-objects">
  <div class="evidence-head">
    <h3><span class="evidence-num">[03]</span>Instantiation & Objects</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Instantiate game objects in GameLevel configuration using Object Literals (data-driven design).</div>
  <p class="evidence-impl">Rather than hard-coding each entity, <code>GameLevel.js</code> holds an array of object literals such as <code>{ class: Player, data: { sprite: '...', spawnX: 100, spawnY: 200 } }</code>. The engine iterates the array and instantiates each entry. This is the 'data-driven design' pattern: the level is data, not code, so designing a new level means editing configuration instead of editing JavaScript &mdash; the same engine handles any level definition.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/oop/inheritance">
  <div class="evidence-head">
    <h3><span class="evidence-num">[04]</span>Inheritance (Basic)</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Create class hierarchy with 2+ levels &mdash; e.g. <code>GameObject → Character → Player</code>.</div>
  <p class="evidence-impl">Our class hierarchy goes three levels deep: <code>GameObject → Character → Player</code>. <code>GameObject</code> provides position, velocity, and the abstract <code>draw</code> skeleton; <code>Character</code> adds animation frames, health, and physics defaults; <code>Player</code> adds keyboard handlers, score, and lives. <code>Enemy</code> and <code>NPC</code> follow the same chain, branching from <code>Character</code>. Three levels deep cleanly satisfies the 2+ requirement and demonstrates real reuse &mdash; the physics code lives in one place, not three.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/oop/method-overriding">
  <div class="evidence-head">
    <h3><span class="evidence-num">[05]</span>Method Overriding</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Override parent lifecycle methods: <code>update()</code>, <code>draw()</code>, <code>handleCollision()</code>.</div>
  <p class="evidence-impl">Each subclass overrides parent lifecycle methods to inject its own behavior. <code>Player.update()</code> reads keyboard flags then calls <code>super.update()</code> for shared physics; <code>Enemy.update()</code> runs patrol AI before delegating to super; <code>Enemy.draw()</code> flips the sprite horizontally when facing left, then calls <code>super.draw()</code> to render. <code>handleCollision()</code> differs entirely per class &mdash; the player loses health on enemy contact, the enemy reverses direction. This is polymorphism in action: same method name, different behavior per type.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/oop/constructor-chaining">
  <div class="evidence-head">
    <h3><span class="evidence-num">[06]</span>Constructor Chaining</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Use <code>super(data, gameEnv)</code> to chain constructors and forward initialization data.</div>
  <p class="evidence-impl">Every subclass constructor calls <code>super(data, gameEnv)</code> as its first statement. This lets the base class handle the universal setup &mdash; registering position, loading the sprite, binding the GameEnv reference &mdash; while the subclass adds its own fields after. Skipping <code>super()</code> would mean rebuilding all that boilerplate in every subclass and would also throw a <code>ReferenceError</code> when the subclass tries to use <code>this</code> before <code>super</code> returns.</p>
</a>
</div>

<div class="category">
<h2>// Control Structures</h2>

<div class="category-intro">
<p>Control structures are how the game makes decisions and repeats actions frame after frame. Loops drive the animation; without them, the game would be a static image. Conditionals drive every choice the game makes; without them, every play-through would be identical. Together they convert raw player input and engine ticks into actual gameplay.</p>
<p>In our game, control structures execute at 60 frames per second. The game loop iterates over every active object each frame, conditionals inside each object's <code>update()</code> decide what to do based on its current state, and nested conditionals handle the corner cases that flat <code>&amp;&amp;</code> chains can't express cleanly.</p>
</div>

<a class="evidence" href="https://chetan-coder.github.io/portfolioC/control/iteration">
  <div class="evidence-head">
    <h3><span class="evidence-num">[01]</span>Iteration</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Use <code>for</code>, <code>forEach</code>, and <code>while</code> loops for game object arrays and animation frames.</div>
  <p class="evidence-impl">Loops drive every frame of the game. We use <code>forEach</code> to call <code>update()</code> on every entity in the <code>gameObjects</code> array (<code>gameObjects.forEach(obj => obj.update())</code>), classic <code>for</code> loops to scan platform arrays during collision checks, and <code>while</code> loops in a procedural sprite-preloading routine. Every frame, the engine iterates through dozens of objects &mdash; the game would visibly hitch if any loop were inefficient.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/control/conditionals">
  <div class="evidence-head">
    <h3><span class="evidence-num">[02]</span>Conditionals</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Use <code>if/else</code> branches to implement collision detection and state transitions.</div>
  <p class="evidence-impl"><code>if/else</code> blocks gate nearly every action. They check whether a key is pressed, whether a collision has happened, whether the player is on the ground, whether the level is complete. A typical line from <code>handleCollision</code>: <code>if (collision.direction === 'top') { this.land(); } else if (collision.direction === 'side') { this.bounceBack(); }</code>. Conditionals are how raw events become game logic.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/control/nested-conditions">
  <div class="evidence-head">
    <h3><span class="evidence-num">[03]</span>Nested Conditions</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Multi-level conditionals for complex game logic (e.g. power-up + collision + direction).</div>
  <p class="evidence-impl">Some rules need layered decisions that single-level <code>if/else</code> can't express cleanly. Power-up pickup is the canonical example: an outer <code>if</code> checks whether a collision happened, an inner <code>if</code> checks if the colliding object is the player, and a deeper <code>if</code> checks whether the player ALREADY has that power-up. Flattening all that into compound <code>&amp;&amp;</code> would be unreadable; nesting keeps each decision traceable.</p>
</a>
</div>

<div class="category">
<h2>// Data Types</h2>

<div class="category-intro">
<p>A game is, at its core, a bag of constantly-updating values. Position is numbers, names are strings, paused-or-not is a boolean, all the entities live in an array, and the level configuration is a JSON object. Choosing the right type for each piece of data is what makes the rest of the code readable instead of fighting type coercion later.</p>
<p>In GateGame v2, all five primary data types appear in every frame of execution &mdash; numbers move things, strings label things, booleans decide things, arrays group things, and JSON-style objects configure things.</p>
</div>

<a class="evidence" href="https://chetan-coder.github.io/portfolioC/data-types/numbers">
  <div class="evidence-head">
    <h3><span class="evidence-num">[01]</span>Numbers</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Position, velocity, and score tracking values.</div>
  <p class="evidence-impl">Every spatial and quantitative value is a number: position <code>(x, y)</code>, velocity <code>(vx, vy)</code>, score, lives, health, frame counters, timers. We use integers for things like score and frame index, and floats for subpixel velocity so movement looks smooth instead of jittery. Most updates are simple arithmetic &mdash; <code>this.y += this.vy * deltaTime</code> runs across every object hundreds of times per second.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/data-types/strings">
  <div class="evidence-head">
    <h3><span class="evidence-num">[02]</span>Strings</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Character names, sprite paths, and game state labels.</div>
  <p class="evidence-impl">Strings label everything the game knows by name. Character names like <code>'Eldric'</code> or <code>'Sentinel'</code> are used for dialogue, sprite paths like <code>'assets/sprites/player_run_01.png'</code> are passed to the asset loader, and game state labels like <code>'playing'</code>, <code>'paused'</code>, or <code>'gameover'</code> drive the state machine. Using descriptive state strings makes <code>console.log(this.state)</code> actually meaningful during debugging.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/data-types/booleans">
  <div class="evidence-head">
    <h3><span class="evidence-num">[03]</span>Booleans</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Flags such as <code>isJumping</code>, <code>isPaused</code>, <code>isVulnerable</code>.</div>
  <p class="evidence-impl">Booleans are the on/off flags that drive game behavior: <code>isJumping</code>, <code>isPaused</code>, <code>isVulnerable</code>, <code>isFacingLeft</code>, <code>isOnGround</code>. A clean game state is mostly booleans plus numbers &mdash; they let us write conditions that read like English: <code>if (this.isOnGround &amp;&amp; !this.isJumping) { this.jump(); }</code>. Booleans are also the cheapest values to compare, which matters when checks run 60 times per second.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/data-types/arrays">
  <div class="evidence-head">
    <h3><span class="evidence-num">[04]</span>Arrays</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Game object collections and level data passed to the engine.</div>
  <p class="evidence-impl">Arrays group entities so we can iterate over them. <code>gameObjects[]</code> holds every active entity in the level; <code>platforms[]</code> is a subset used during collision checks; <code>enemies[]</code> is iterated for AI updates. Without arrays, the main loop would be a hard-coded list of variable names that broke any time the level changed. Arrays are also what makes data-driven level config possible: the level is just an array of object literals.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/data-types/objects">
  <div class="evidence-head">
    <h3><span class="evidence-num">[05]</span>Objects (JSON)</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Configuration objects and sprite data &mdash; data-driven GameLevel design.</div>
  <p class="evidence-impl">JSON-style object literals configure everything from sprite metadata (<code>{ src, frames, frameRate, scale }</code>) to NPC dialogue trees to the entire level definition. This is the data-driven design pattern in practice: keep the data in plain objects, let the engine read them, and you can change the game without touching the engine code. The same engine renders any level whose config matches the expected shape.</p>
</a>
</div>

<div class="category">
<h2>// Operators</h2>

<div class="category-intro">
<p>Operators are the verbs of JavaScript &mdash; the symbols that actually do something with the data. Math operators move objects through space, boolean operators gate every choice the game makes, and string operators build all the text and paths the player sees.</p>
<p>In GateGame v2, the physics loop is pure math operators, the collision logic is compound boolean expressions, and the UI/HUD is built from template literals &mdash; all three appear in every gameplay frame.</p>
</div>

<a class="evidence" href="https://chetan-coder.github.io/portfolioC/operators/mathematical">
  <div class="evidence-head">
    <h3><span class="evidence-num">[01]</span>Mathematical</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Operators <code>+ - * /</code> in physics calculations: gravity, velocity, collision response.</div>
  <p class="evidence-impl">Physics is just arithmetic operators applied to position and velocity each frame: <code>this.y += this.vy; this.vy += gravity;</code>. Collision response uses subtraction to nudge objects back out of overlap, multiplication for area calculations, and division for normalizing direction vectors. Modulo <code>%</code> wraps sprite animation frame indices when they exceed the frame count, so the animation loops smoothly.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/operators/string-ops">
  <div class="evidence-head">
    <h3><span class="evidence-num">[02]</span>String Operations</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Template literals and concatenation for sprite paths and on-screen text display.</div>
  <p class="evidence-impl">Template literals like <code>`${this.name}: ${this.health}`</code> build the HUD overlay text every frame. Concatenation (<code>'assets/sprites/' + this.name + '_run_' + frameIndex + '.png'</code>) assembles sprite paths from base directories plus dynamic parts. Template literals are easier to read than long concatenation chains, so we lean on them everywhere a string is interpolated.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/operators/boolean-expr">
  <div class="evidence-head">
    <h3><span class="evidence-num">[03]</span>Boolean Expressions</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Logical <code>&amp;&amp;</code>, <code>||</code>, <code>!</code> for compound conditions in game logic.</div>
  <p class="evidence-impl">Compound conditions gate every meaningful action. <code>&amp;&amp;</code> combines requirements (<code>if (collision &amp;&amp; !this.isVulnerable &amp;&amp; enemy.isAlive)</code>), <code>||</code> handles alternatives (<code>if (key === 'ArrowUp' || key === 'w')</code>), and <code>!</code> flips state (<code>if (!this.isPaused)</code>). These short-circuit, so we order the operands with the cheapest check first as a small performance habit &mdash; the loop runs thousands of times per second.</p>
</a>
</div>

<div class="category">
<h2>// Input / Output</h2>

<div class="category-intro">
<p>Input/Output is everything the game does that crosses a boundary &mdash; reading the player's keyboard, drawing pixels to the canvas, fetching leaderboard scores from the backend, sending the player's words to the NPC AI. These are the operations where the game touches the outside world (the user, the screen, the network).</p>
<p>In our game, I/O splits cleanly into two halves: <strong>local I/O</strong> (keyboard, canvas) which is synchronous and runs every frame, and <strong>remote I/O</strong> (Leaderboard + NPC AI) which is asynchronous and must not block the game loop. Both halves use modern JavaScript APIs and both need correct error handling.</p>
</div>

<a class="evidence" href="https://chetan-coder.github.io/portfolioC/io/keyboard">
  <div class="evidence-head">
    <h3><span class="evidence-num">[01]</span>Keyboard Input</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Arrow keys, Space, and WASD controls using event listeners.</div>
  <p class="evidence-impl">Event listeners on <code>keydown</code> and <code>keyup</code> set boolean flags on the player object: <code>this.keys.right = true</code> when ArrowRight is pressed, false when released. Reading those flags inside <code>update()</code> &mdash; instead of moving directly in the handler &mdash; gives smooth simultaneous movement (e.g. holding right + jump at the same time). Arrow keys, WASD, Space, and E (for NPC interaction) are all wired up this way.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/io/canvas">
  <div class="evidence-head">
    <h3><span class="evidence-num">[02]</span>Canvas Rendering</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Draw sprites, backgrounds, and platforms using the Canvas API.</div>
  <p class="evidence-impl">Every pixel goes through the Canvas 2D API. Each frame we clear the canvas, draw the parallax background, then iterate <code>gameObjects</code> and call each one's <code>draw(ctx)</code> method. <code>ctx.drawImage</code> renders sprites, <code>ctx.fillText</code> renders HUD numbers, and during debug mode <code>ctx.strokeRect</code> outlines hit boxes. Layer order matters &mdash; background first, then platforms, then characters, then HUD on top.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/io/gameenv">
  <div class="evidence-head">
    <h3><span class="evidence-num">[03]</span>GameEnv Configuration</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Set canvas size, difficulty levels, and game settings.</div>
  <p class="evidence-impl">A central <code>GameEnv</code> object holds canvas width/height, gravity, difficulty multipliers, and a debug flag. Every game object holds a reference to it (passed via <code>super(data, gameEnv)</code>) and queries it during <code>update()</code> &mdash; for example, an enemy reads <code>gameEnv.difficulty</code> to scale its damage. Centralizing config in one object means a single edit changes the whole level's behavior.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/io/api">
  <div class="evidence-head">
    <h3><span class="evidence-num">[04]</span>API Integration</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Implement Leaderboard API (POST/GET scores) and NPC AI interactions.</div>
  <p class="evidence-impl">The Leaderboard API uses <code>POST</code> to submit scores when a level is completed and <code>GET</code> to retrieve the top 10 for display on the leaderboard screen. The NPC AI integration POSTs the player's typed message to an AI backend and uses the response text as in-game dialogue. Both integrations include explicit error handling so a network failure doesn't crash the game &mdash; the leaderboard falls back to localStorage and the NPC falls back to canned dialogue.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/io/async">
  <div class="evidence-head">
    <h3><span class="evidence-num">[05]</span>Asynchronous I/O</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Use <code>async/await</code> or Promises for API calls without blocking the game loop.</div>
  <p class="evidence-impl">Every fetch call uses <code>async/await</code> rather than chained <code>.then()</code>, which keeps the code readable. The game loop never blocks on the network &mdash; the leaderboard screen renders 'loading...' until the Promise resolves, and the player can keep playing the main level while a leaderboard request is in flight. Avoiding blocking I/O is what keeps the framerate at 60fps even on a slow connection.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/io/json-parsing">
  <div class="evidence-head">
    <h3><span class="evidence-num">[06]</span>JSON Parsing</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Parse API responses for leaderboard data and AI responses.</div>
  <p class="evidence-impl"><code>response.json()</code> parses fetch responses into native JavaScript objects. The leaderboard returns an array like <code>[{ user: 'chetan', score: 4200 }, ...]</code> which we iterate and render row by row. The NPC AI returns <code>{ message: '...', emotion: '...' }</code> which we destructure: <code>const { message, emotion } = await response.json()</code>. Both responses are validated before being used to avoid runtime errors on a malformed payload.</p>
</a>

<h2>// Debugging</h2>

<div class="category-intro">
<p>When a game breaks &mdash; and games always break &mdash; Chrome DevTools is the diagnostic toolkit. Each tab is best for a different class of bug: Console for state inspection, Sources for step-by-step execution, Network for API issues, Application for storage and session problems, Elements for layout and rendering.</p>
<p>I used every one of these tabs at some point during development. The most useful technique turned out to be hit-box visualization &mdash; drawing the collision rectangles directly onto the canvas during debug mode &mdash; because some bugs only become obvious when you can actually see the geometry.</p>
</div>

<a class="evidence" href="https://chetan-coder.github.io/portfolioC/debugging/console">
  <div class="evidence-head">
    <h3><span class="evidence-num">[01]</span>Console Debugging</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Use <code>console.log</code> to track game state, variables, and method calls.</div>
  <p class="evidence-impl">Strategic <code>console.log</code> calls inside <code>update()</code> and the collision methods print state every frame during dev. <code>console.warn</code> flags unexpected branches and <code>console.error</code> is used for caught exceptions. All console output is guarded behind <code>gameEnv.debug</code> so it's silent in production. This is the fastest way to see what variables are doing in real time without pausing execution.</p>
</a>
<a class="evidence" href="https://chetan-coder.github.io/portfolioC/debugging/hitbox">
  <div class="evidence-head">
    <h3><span class="evidence-num">[02]</span>Hit Box Visualization</h3>
    <span class="evidence-arrow">→</span>
  </div>
  <div class="evidence-req">Evidence required: Draw / visualize collision boundaries to refine detection during development.</div>
  <p class="evidence-impl">When debug mode is on, every object renders a colored stroke around its collision rectangle on top of its sprite. This is how I caught a bug where the player's hit box was 2px wider than the sprite, causing false enemy collisions on the right edge. Adjusting the rect fixed it &mdash; but I wouldn't have noticed the visual offset without literally drawing the boxes on screen.</p>

<a class="evidence" href="https://chetan-coder.github.io/portfolioC/debugging/sources">

  <p class="evidence-impl">I set breakpoints inside <code>handleCollision()</code> and stepped through frame by frame to find the exact moment a wrong direction value was being returned. Stepping through code is invaluable when a value is wrong but <code>console.log</code> can't pinpoint when it changed &mdash; pausing execution and inspecting the call stack reveals the sequence that led there.</p>

<a class="evidence" href="https://chetan-coder.github.io/portfolioC/debugging/network">
